# HOW CAN YOU CHECK YOUR RANDOM VARIABLE IS NORMAL DISTRIBUTION OR NOT
**************************************************************************************

## HERE IS WHERE QQ PLOTS PLAYS A ROLE 
## LETS SEE HOW

```python
import numpy as np
norm = np.random.normal(loc = 0.0, scale = 1.0 , size = 1000)
import seaborn as sns
sns.distplot(norm)
```

# LETS TAKE REAL DATA SET

```python
import pandas as pd
df = pd.read_csv('train (2).csv')                     #titanic dataset
age_values = df.Age.dropna()
sns.distplot(age_values)                              #lets check whether age is normally distributed or not
```

```python
from scipy.stats import probplot
```

```python
import pylab
g = probplot(age_values,dist='norm',plot=pylab)
pylab.show(g)                                # age values are more or less normally distributed
```
